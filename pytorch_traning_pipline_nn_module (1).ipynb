# **Scaling**

scaler = StandardScaler(): This line creates an instance of the StandardScaler class. StandardScaler standardizes features by removing the mean and scaling to unit variance. This results in a distribution with a mean of 0 and a standard deviation of 1.

x_train = scaler.fit_transform(x_train): This is applied to your training data (x_train).

fit: This method calculates the mean and standard deviation of each feature in x_train.
transform: This method then uses the calculated means and standard deviations to scale the x_train data.
x_test = scaler.transform(x_test): This is applied to your test data (x_test).

transform: Crucially, for the test set, we only use the transform method. This means it applies the same mean and standard deviation that were calculated from the training data to scale the test data. We do not use fit_transform on x_test because we want to prevent data leakage from the test set into the training process.
**Why do we do this?**

Feature scaling is an important preprocessing step for many machine learning algorithms, especially those that are sensitive to the magnitude of feature values. Here's why it's often done:

**Algorithm Performance**: Algorithms like Logistic Regression, Support Vector Machines (SVMs), k-Nearest Neighbors (k-NN), and Neural Networks are often sensitive to the scale of features. If one feature has a much larger range of values than another, the algorithm might disproportionately weigh that feature.
**Faster Convergence**: For gradient-descent-based algorithms, scaling can lead to faster convergence during training.
**Preventing Domination**: Features with larger values might dominate the distance calculations or cost functions, even if they are not more important predictors. Scaling ensures that all features contribute equally to the distance metrics or cost functions.
In essence, it brings all your features to a similar scale, which helps many models perform better and more consistently.

# **Numpy arrays to pytorch tensors**

```python
x_train_tensor = torch.from_numpy(x_train).float()
x_test_tensor = torch.from_numpy(x_test).float()
y_train_tensor = torch.from_numpy(y_train).float()
y_test_tensor = torch.from_numpy(y_test).float()
```

# Defining **Model**

```python
import torch
import torch.nn as nn
class MySimpeNN(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

# **Important Parameters**

```python
learning_rate = 0.1
epochs = 25
```

# **Loss function**

```python
loss_function = nn.BCELoss()
```

```python
type(loss_function)
```

```python
#define optimizer
```

## **Traning pipline of nn module**

```python
# create model
model = MySimpeNN(x_train_tensor.shape[1])

#define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

#define loop
for epoch in range(epochs):
  #forward pass
  y_pred = model(x_train_tensor)

  #loss calculation
  loss = loss_function(y_pred, y_train_tensor.view(-1,1))

  #zero gradients clear them before the backward pass it is best and then update the parameters
  optimizer.zero_grad()

  #backwardpass
  loss.backward()

  #update weights
  optimizer.step()

  #print loss in each epoch
  print(f"Epoch {epoch}/{epoch}, Loss: {loss.item()}")
```

```python
model.linear.weight
```

# **Evaluation**

```python
with torch.no_grad():
  y_pred = model.forward(x_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f"Accuracy: {accuracy}")
```

```python
#create model class
import torch
import torch.nn as nn


class Model(nn.Module):
  def __init__(self, num_features):
#invoking the parent class(nn.Module) constructor through super() keyword
    super().__init__()
    # 3 in the parameter is the number of neuron and num_features is the number of inputs
    self.linear1 = nn.Linear(num_features, 3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3,1)
    self.sigmoid = nn.Sigmoid()

    #forward pass
  def forward(self, features):
      out = self.linear1(features) #z = wx+b
      out = self.relu(out)
      out = self.linear2(out)
      out = self.sigmoid(out)
      return out
```

# Sequential **Container**

```python
#create model class
import torch
import torch.nn as nn


class Model(nn.Module):
  def __init__(self, num_features):
#invoking the parent class(nn.Module) constructor through super() keyword
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid(),
    )

    #forward pass
  def forward(self, features):
      out  = self.network(features)
      return out
```

### What is `nn.Sequential`?
`torch.nn.Sequential` is a module that contains other modules (layers) and passes the input through them in the order they are added. It's essentially a linear stack of neural network layers. When you call the `forward` method on an `nn.Sequential` module, the input is passed to the first layer, its output becomes the input to the second layer, and so on, until the last layer's output is returned.

### How it Helps:

1.  **Simplicity and Readability**: It makes your code cleaner and easier to understand, especially for models with a straightforward, feed-forward architecture. Instead of defining each layer and explicitly chaining them in the `forward` method, you list them once.

2.  **Modularity**: Each layer within `nn.Sequential` is an `nn.Module` itself. This means you can combine different types of layers (linear, convolutional, activation functions, pooling, etc.) seamlessly.

3.  **Automatic `forward` pass**: You don't need to write the `forward` method yourself when using `nn.Sequential` for simple, sequential models. PyTorch handles the data flow through the layers automatically.

4.  **Parameter Management**: All parameters (weights and biases) of the layers within `nn.Sequential` are automatically registered as parameters of the `Sequential` module itself. This simplifies optimization, as you only need to optimize the parameters of the `Sequential` module.

5.  **Easy Inspection**: As you saw in the previous example, you can easily access individual layers and their parameters by indexing the `nn.Sequential` object (e.g., `model.network[0]` for the first layer, `model.network[2]` for the second linear layer, etc.).

### When to Use It (and When Not To):

*   **Ideal for**: Models with a clear, linear data flow, where the output of one layer directly feeds into the next. Many common architectures, like simple fully connected networks or stacked convolutional layers, fit this pattern.

*   **Less Suitable for**: Models with more complex architectures, such as those with multiple inputs/outputs, skip connections (e.g., ResNets), branching paths, or custom logic that depends on the intermediate outputs of multiple layers. For these, defining a custom `nn.Module` with an explicit `forward` method is necessary.
