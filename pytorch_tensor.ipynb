```python
b =torch.rand(2,3,4) #first number 2 represents the number of blocks
b
```

```python
#permute  is used to rearrange the dimensions of a tensor. When you call b.permute(2,0,1) on a tensor b with shape (2,3,4)
b.permute(2,0,1)
```

```python
b.permute(2,0,1).shape
```

```python
#unsqueeze operation is used to add a new dimension of size 1 at the specified position (index 0 in this case) of the tensor. This is commonly used in deep learning to add a batch dimension to a single image or data sample.
#image size
c = torch.rand(226,226,3)
c.unsqueeze(0).shape
```

```python
#The squeeze(0) operation removes dimensions of size 1 from a tensor.d.squeeze(0) removes this dimension of size 1 at index 0.
#The resulting shape is (20), effectively flattening the tensor into a 1D vector.
d = torch.rand(1,20)
d.squeeze(0).shape
```
